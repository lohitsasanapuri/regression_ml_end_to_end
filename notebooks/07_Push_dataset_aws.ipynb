```python
import boto3
from pathlib import Path
```

```python
# --- Configuration Template ---
bucket_name = "your-bucket-name"
region = "your-region"
access_key = "your-access-key"
secret_key = "your-secret-key"
```

```python
PROJECT_ROOT  = Path.cwd().parent
local_data_dir = PROJECT_ROOT / "data" / "processed"
local_model_dir = PROJECT_ROOT / "models"
```

```python
s3 = boto3.client(
    "s3",
    region_name=region,
    aws_access_key_id=access_key,
    aws_secret_access_key=secret_key
)

# Upload processed data files to S3
def upload_file_to_s3(local_path: Path, s3_key : str):
    if not local_path.exists():
        print(f"File {local_path} does not exist.")
        return
    s3.upload_file(str(local_path), bucket_name, s3_key)
    print(f"Uploaded {local_path} to s3://{bucket_name}/{s3_key}")
    
```

```python
# ---- Upload required datasets ----
upload_file_to_s3(local_data_dir / "test_fe.csv", "processed/test_fe.csv")
upload_file_to_s3(local_data_dir / "test_cleaned.csv", "processed/test_cleaned.csv")
upload_file_to_s3(local_data_dir / "train_fe.csv", "processed/train_fe.csv")

# ---- Upload model ----
upload_file_to_s3(local_model_dir / "xgb_model.pkl", "models/xgb_model.pkl")
```
