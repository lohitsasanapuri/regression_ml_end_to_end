```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import  StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
train_df = pd.read_csv("../data/processed/train_fe.csv")
val_df = pd.read_csv("../data/processed/val_fe.csv")
```

* Dropping High Correlated Columns from the Train and Val

```python
target = "price"
x_train = train_df.drop(columns=[target])
y_train = train_df[target]
x_val = val_df.drop(columns=[target])
y_val = val_df[target]
```

```python
# Standardize the features
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_val_scaled = scaler.transform(x_val)
```

## Simple Linear Regression Model

```python
lr = LinearRegression()
lr.fit(x_train_scaled, y_train)
y_pred = lr.predict(x_val_scaled)

print("MAE:", mean_absolute_error(y_val, y_pred))   
print("MSE:", mean_squared_error(y_val, y_pred))
print("R2:", r2_score(y_val, y_pred))
```

## Ridge Regresssion Model

```python
ridge_Lr = Ridge(alpha=1.0)
ridge_Lr.fit(x_train_scaled, y_train)
y_pred_ridge = ridge_Lr.predict(x_val_scaled)
print("Ridge Regression")

print("MAE:", mean_absolute_error(y_val, y_pred_ridge))
print("MSE:", mean_squared_error(y_val, y_pred_ridge))
print("R2:", r2_score(y_val, y_pred_ridge))
```

## Lasso Regression Model

```python
lasso_lr = Lasso(alpha=0.1)
lasso_lr.fit(x_train_scaled, y_train)
y_pred_lasso = lasso_lr.predict(x_val_scaled)

print("Lasso Regression")
print("MAE:", mean_absolute_error(y_val, y_pred_lasso))
print("MSE:", mean_squared_error(y_val, y_pred_lasso))
print("R2:", r2_score(y_val, y_pred_lasso))
```

## ElesticNet Regression Model

```python
elasticnet_lr = ElasticNet(alpha=0.1, l1_ratio=0.5)
elasticnet_lr.fit(x_train_scaled, y_train)
y_pred_elasticnet = elasticnet_lr.predict(x_val_scaled)
print("ElasticNet Regression")
print("MAE:", mean_absolute_error(y_val, y_pred_elasticnet))
print("MSE:", mean_squared_error(y_val, y_pred_elasticnet)) 
print("R2:", r2_score(y_val, y_pred_elasticnet))
```
